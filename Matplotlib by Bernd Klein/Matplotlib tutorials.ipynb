# Overview of matplotlib

###  Simple line plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot([-1, -4.5, 16, 23])
plt.show()
```

```python
plt.plot([-1, -4.5, 16, 23], 'ob')
plt.plot([-1, -4.5, 16, 23])
plt.show()
```

```python
days = range(1, 9)
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]

fig, ax = plt.subplots()
ax.plot(days, celsius_values)
ax.set(xlabel='Day',
       ylabel='Temperature in Celsius',
       title='Temperature Graph')
plt.show()
```

### Multiple plots in one graph

```python
days = list(range(1, 9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]

fig, ax = plt.subplots()
ax.set(xlabel='Day',
       ylabel='Temperature in Celsius',
       title='Temperature Graph')

ax.plot(days, celsius_min)
ax.plot(days, celsius_min, "oy")
ax.plot(days, celsius_max)
ax.plot(days, celsius_max, "or")
plt.show()
```

### Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np

years = [str(year) for year in range(2010, 2021)]
visitors = (1241, 50927, 162242, 222093, 665004, 2071987, 2460407, 3799215, 5399000, 5474016, 6003672)

plt.bar(years, visitors, color='grey')
plt.xlabel("Years")
plt.ylabel("Visitors")
plt.title("Bar chart")

plt.show()
```

### Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
gaussian_numbers = np.random.normal(size=1000)

plt.hist(gaussian_numbers, bins=20)
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### Scatter plots

```python
x = np.arange(0, 11)
y1 = np.random.randint(2, 7, (11,))
y2 = np.random.randint(9, 14, (11,))
y3 = np.random.randint(15, 25, (11,))

plt.scatter(x, y1)
plt.scatter(x, y2, marker="*", color='r')
plt.scatter(x, y3, marker="x", color='y')
plt.title("Scatter plot example")
plt.show()
```

### Stack plot

```python
indexes = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [23, 42, 33, 43, 8, 44, 43, 18, 21]
y2 = [9, 31, 25, 14, 17, 17, 42, 22, 28]
y3 = [18, 29, 19, 22, 18, 16, 13, 32, 21]

plt.stackplot(indexes, y1, y2, y3)
plt.title("Stack plot")
plt.show()
```

### Pie charts

```python
labels = ['C', 'python', 'Java', 'C++', 'C#']
sizes = [13.38, 11.87, 11.74, 7.81, 4.41]
explode = (0, 0.1, 0, 0, 0) #explodes the 2nd slice.

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, shadow=True, autopct='%.1f%%', startangle=0)
ax.axis('equal')
plt.title("Pie chart")
plt.show()
    
```

```python
labels = ['C', 'python', 'Java', 'C++', 'C#', 'others']
sizes = [13.38, 11.87, 11.74, 7.81, 4.41]
sizes.append(100-sum(sizes))
explode = (0, 0.1, 0, 0, 0, 0) #explodes the 2nd slice.

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, shadow=True, autopct='%.1f%%', startangle=0)
ax.axis('equal')
plt.title("Pie chart")
plt.show()
    
```

### The plot function

```python
days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]

fig, ax = plt.subplots()

#formatting the graph
ax.set( xlabel='Days',
        ylabel = 'Temperature in Celsius',
        title = 'Temperature Graph')
ax.plot(days, celsius_min,
        days, celsius_min, 'oy',
        days, celsius_max,
        days, celsius_max, 'or')

plt.show()
print(ax.axis())
```

### Checking and defining range of axes

```python
import matplotlib.pyplot as plt

days = list(range(1, 9))
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1] 

fig, ax = plt.subplots()
ax.plot(days, celsius_values)

ax.set(xlabel='Day',
       ylabel='Temperature in Celsius',
       title='Temperature Graph')

print(ax.axis())
xmin, xmax, ymin, ymax = 0, 10, 14, 45
ax.axis([xmin, xmax, ymin, ymax])
print(ax.axis())
plt.show()
```

### Linspace to define x values

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
f1 = 3 * np.sin(x)
f2 = np.sin(2*x)
f3 = 0.3 * np.sin(x)

fig, ax = plt.subplots()


startx, endx = -2 * np.pi - 0.1, 2*np.pi + 0.1 
starty, endy = -3.1, 3.1
ax.axis([startx, endx, starty, endy])
ax.plot(x, f1, x, f2, x, f3)
#plotting discrete points
#ax.plot(x, f1, 'ro', x, f2, 'bx')
plt.show()
```

### Changing the line style

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2 * np.pi, 50, endpoint=True)
f1 = 3 * np.sin(x)
f2 = np.sin(2*x)
f3 = 0.3 * np.sin(x)
f4 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, f1, color="blue", linewidth=2.5, linestyle="-")
ax.plot(x, f2, color="red", linewidth=1.5, linestyle="--")
ax.plot(x, f3, color="green", linewidth=2, linestyle=":")
ax.plot(x, f4, color="grey", linewidth=2, linestyle="-.")
plt.show()
```
