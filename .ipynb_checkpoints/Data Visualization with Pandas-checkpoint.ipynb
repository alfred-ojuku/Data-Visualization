## Line plots in Pandas
### Series

```python
import pandas as pd

data = [100, 120, 140, 180, 200, 210, 214]
s = pd.Series(data, index=range(len(data)))
s.plot()
```

It is possible to suppress the usage of the index by setting the keyword parameter "use_index" to False.

```python
s.plot(use_index=False)
```

I will experiment now with a Series which has an index consisting of alphabetical values.

```python
fruits = ['apples', 'oranges', 'cherries', 'pears']
quantities = [20, 33, 52, 10]
S = pd.Series(quantities, index=fruits)
S.plot()
```

### DataFrame

```python
import pandas as pd

cities = {"name": ["London", "Berlin", "Madrid", "Rome",
                   "Paris", "Vienna", "Bucharest", "Hamburg",
                   "Budapest", "Warsaw", "Barcelona",
                   "Munich", "Milan"],
          "population": [8615246, 3562166, 3165235, 2874038,
                         2273305, 1805681, 1803425, 1760433,
                         1754000, 1740119, 1602386, 1493900,
                         1350680],
          "area" : [1572, 891.85, 605.77, 1285,
                    105.4, 414.6, 228, 755,
                    525.2, 517, 101.9, 310.4,
                    181.8]}
          
df = pd.DataFrame(cities, columns=['population', 'area'], index=cities['name'])
df
```

```python
df['area'] *= 1000
df.plot()
```

The above plot is not coming up to my expectations, because not all the city names appear on the x axis.I can change this by defining the xticks explicitly with "range(len((city_frame.index))"

```python
df.plot(xticks=range(len(df.index)), use_index=True)
```

Now, I have a new problem. The city names are overlapping.

```python
df.plot(xticks=range(len(df.index)), use_index=True, rot=45)
```

### Using Twin axes
I multiplied the area column by 1000 to get a proper output. Instead of this, I could have used twin axes.

```python
f = pd.DataFrame(cities, columns=['population', 'area'], index=cities['name'])
f
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.suptitle("City statistics")
ax.set(xlabel='population',
       ylabel='Cities')

ax2 = ax.twinx()
ax2.set(ylabel='Area')

f['population'].plot(ax=ax, style="b-", xticks=range(len(f.index)), rot=90)
f['area'].plot(ax=ax2, style='g-', rot=90)
ax.legend(['population number'], loc=2)
ax2.legend(loc=1)
plt.show()
```

```python
import matplotlib.pyplot as plt


ax1= f["population"].plot(style="b-", xticks=range(len(f.index)), use_index=True, rot=90)
ax2 = ax1.twinx()
f["area"].plot(ax=ax2, style="g-", use_index=True, rot=90)
ax1.legend(loc = (.7,.9), frameon = False)
ax2.legend( loc = (.7, .85), frameon = False)
plt.show()
```

### Multiple Y axis

```python
f['density'] = f['population'] / f['area']
f
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.suptitle("City Statistics")
ax.set_ylabel("Population")
ax.set_xlabel("Citites")

ax_area, ax_density = ax.twinx(), ax.twinx()
ax_area.set_ylabel("Area")
ax_density.set_ylabel("Density")

rspine = ax_density.spines['right']
rspine.set_position(('axes', 1.25))
ax_density.set_frame_on(True)
ax_density.patch.set_visible(False)
fig.subplots_adjust(right=0.75)

f["population"].plot(ax=ax, style="b-", xticks=range(len(f.index)), use_index=True, rot=90)
f["area"].plot(ax=ax_area, style="g-", use_index=True, rot=90)
f["density"].plot(ax=ax_density, style="r-", use_index=True, rot=90)
plt.show()
```
